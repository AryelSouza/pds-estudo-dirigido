```python
# Parte 3  Implementao de Filtros Discretos no Tempo

import numpy as np
import matplotlib.pyplot as plt
from scipy import signal

# Sinal de entrada: senoide com rudo
fs = 100  # Hz
t = np.linspace(0, 1, fs, endpoint=False)
x = np.sin(2 * np.pi * 5 * t) + 0.5 * np.random.randn(len(t))

# Filtro passa-baixas IIR (Butterworth de ordem 4, cutoff 10 Hz)
b, a = signal.butter(4, 10, btype='low', fs=fs)

# Aplicar o filtro
y = signal.lfilter(b, a, x)

# Plot
plt.figure(figsize=(12, 5))
plt.plot(t, x, label='Sinal com rudo')
plt.plot(t, y, label='Sinal filtrado', linewidth=2)
plt.title('Filtragem de sinal ruidoso com filtro IIR Butterworth')
plt.xlabel('Tempo [s]')
plt.ylabel('Amplitude')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
#Filtro passa-altas
fs = 500
t = np.linspace(0, 1, fs, endpoint=False)
x = np.sin(2 * np.pi * 10 * t) + np.sin(2 * np.pi * 100 * t)  # Baixa e alta frequncia

# Filtro passa-altas (corta 10Hz)
b, a = signal.butter(4, 30, btype='high', fs=fs)
y = signal.lfilter(b, a, x)

plt.figure(figsize=(10, 4))
plt.plot(t, x, label='Sinal original')
plt.plot(t, y, label='Sinal filtrado (passa-altas)', linewidth=2)
plt.legend()
plt.grid(True)
plt.title("Filtro passa-altas")
plt.xlabel("Tempo")
plt.ylabel("Amplitude")
plt.tight_layout()
plt.show()
```
