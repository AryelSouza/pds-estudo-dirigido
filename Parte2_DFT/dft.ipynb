```python
import numpy as np
import matplotlib.pyplot as plt

# Sinal senoidal
N = 64
n = np.arange(N)
f = 5  # frequência do sinal
x = np.sin(2 * np.pi * f * n / N)

# Cálculo da DFT usando FFT
X = np.fft.fft(x)
freqs = np.fft.fftfreq(N, d=1)

# Gráfico do espectro de magnitude e fase
plt.figure(figsize=(12, 6))

plt.subplot(2, 1, 1)
plt.stem(freqs, np.abs(X))
plt.title("Espectro de Magnitude da DFT")
plt.xlabel("Frequência normalizada")
plt.ylabel("|X[k]|")
plt.grid(True)

plt.subplot(2, 1, 2)
plt.stem(freqs, np.angle(X))
plt.title("Espectro de Fase da DFT")
plt.xlabel("Frequência normalizada")
plt.ylabel("Fase [rad]")
plt.grid(True)

plt.tight_layout()
plt.show()
```
