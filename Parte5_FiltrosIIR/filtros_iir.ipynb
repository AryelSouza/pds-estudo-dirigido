```python
# Parte 5  Projeto de Filtros IIR com filtro Butterworth

import numpy as np
import matplotlib.pyplot as plt
from scipy import signal

# Sinal com rudo
fs = 1000  # Hz
t = np.linspace(0, 1, fs, endpoint=False)
x = np.sin(2 * np.pi * 50 * t) + 0.5 * np.sin(2 * np.pi * 300 * t)

# Projeto do filtro IIR passa-baixas (Butterworth de ordem 4)
order = 4
cutoff = 100  # Hz
b, a = signal.butter(order, cutoff, btype='low', fs=fs)

# Aplicao do filtro
y = signal.lfilter(b, a, x)

# Plot
plt.figure(figsize=(12, 5))
plt.plot(t, x, label='Sinal original')
plt.plot(t, y, label='Sinal filtrado (IIR)', linewidth=2)
plt.title('Filtro IIR Butterworth passa-baixas')
plt.xlabel('Tempo [s]')
plt.ylabel('Amplitude')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
#Filtro Chebyshev Tipo I
fs = 500
t = np.linspace(0, 1, fs, endpoint=False)
x = np.sin(2 * np.pi * 20 * t) + np.sin(2 * np.pi * 200 * t)

# Filtro Chebyshev Tipo I passa-baixas
b, a = signal.cheby1(N=4, rp=1, Wn=50, btype='low', fs=fs)
y = signal.lfilter(b, a, x)

plt.figure(figsize=(10, 4))
plt.plot(t, x, label='Sinal original')
plt.plot(t, y, label='Chebyshev Tipo I', linewidth=2)
plt.title("Filtro IIR Chebyshev Tipo I")
plt.xlabel("Tempo")
plt.ylabel("Amplitude")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
