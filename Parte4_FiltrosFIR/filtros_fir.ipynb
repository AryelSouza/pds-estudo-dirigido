```python

# Parte 4  Projeto de Filtros FIR com janela de Hamming

import numpy as np
import matplotlib.pyplot as plt
from scipy import signal

# Sinal de entrada: senoide com rudo de alta frequncia
fs = 1000
t = np.linspace(0, 1, fs, endpoint=False)
x = np.sin(2 * np.pi * 50 * t) + 0.5 * np.sin(2 * np.pi * 250 * t)

# Projeto do filtro FIR passa-baixas com janela de Hamming
numtaps = 51  # nmero de coeficientes
cutoff = 100  # frequncia de corte em Hz
b = signal.firwin(numtaps=numtaps, cutoff=cutoff, fs=fs, window='hamming')

# Aplicar o filtro
y = signal.lfilter(b, [1.0], x)

# Plot dos sinais
plt.figure(figsize=(12, 5))
plt.plot(t, x, label='Sinal original')
plt.plot(t, y, label='Sinal filtrado (FIR)', linewidth=2)
plt.title('Filtro FIR passa-baixas com janela de Hamming')
plt.xlabel('Tempo [s]')
plt.ylabel('Amplitude')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
